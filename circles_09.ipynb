```python
import random
import math
import matplotlib.pyplot as plt
n= random.uniform(-1,1)
p= random.uniform(-1, 1)

def is_in(x,y,r,n,p):
    print(n,p)
    return math.sqrt(((n-x)**2)+((p-y)**2))<=r

x,y,r=0,0,1
is_in(x,y,r,n,p)


compte=0
for i in range(1000):
    n= random.uniform(-1, 1)
    p= random.uniform(-1, 1)
    if is_in(x,y,r,n,p):
        compte+=1
z=4*compte/1000
print(z)

points_x, points_y = [], []

for i in range(1000):
    n= random.uniform(-1, 1)
    p= random.uniform(-1, 1)
    points_x.append(n)
    points_y.append(p)


fig, a = plt.subplots()
circle = plt.Circle((x,y), r, fill=False, color='blue')
a.add_patch(circle)
a.scatter(points_x, points_y, s=5, color='red')
a.set_aspect('equal')
plt.show()


    



```
